### Outlier Detection using Interquartile Range (IQR)

Outliers are data points that significantly differ from other observations. They can often be caused by measurement errors or simply be indicative of a long-tail distribution. The Interquartile Range (IQR) method is a robust way to identify them.

**How it works:**
1.  **Calculate Q1 and Q3**: These are the 25th and 75th percentiles of the data, respectively.
2.  **Calculate IQR**: `IQR = Q3 - Q1`.
3.  **Define Bounds**: Data points falling outside these bounds are considered outliers:
    *   Lower Bound: `Q1 - 1.5 * IQR`
    *   Upper Bound: `Q3 + 1.5 * IQR`

I will apply this method to the 'Price' column of the `df` DataFrame.

```python
import pandas as pd
import os

# Define the path to the existing CSV file in Colab's sample data
input_csv_path = '/content/sample_data/my_sample_data.csv'

df = None # Initialize df to None

# Load the DataFrame 'df'
try:
    df = pd.read_csv(input_csv_path)
    print(f"DataFrame loaded from '{input_csv_path}':")
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file '{input_csv_path}' was not found. Please ensure it exists by running the cell that creates it (e.g., cell `210a85c9`) first.")
except Exception as e:
    print(f"An error occurred while loading df: {e}")

# Only proceed if df was successfully loaded
if df is not None:
    print("Original DataFrame head:")
    display(df.head())

    # --- Outlier Detection using IQR on 'Price' column ---

    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df['Price'].quantile(0.25)
    Q3 = df['Price'].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"\nQ1 (25th percentile) for Price: {Q1:.2f}")
    print(f"Q3 (75th percentile) for Price: {Q3:.2f}")
    print(f"IQR (Interquartile Range) for Price: {IQR:.2f}")
    print(f"Lower Bound for Outliers: {lower_bound:.2f}")
    print(f"Upper Bound for Outliers: {upper_bound:.2f}")

    # Identify outliers
    outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

    if not outliers.empty:
        print("\nIdentified Outliers in 'Price' column:")
        display(outliers)
    else:
        print("\nNo outliers found in the 'Price' column using the IQR method.")
else:
    print("\nDataFrame 'df' was not loaded. Skipping outlier detection.")
```

```python
print("Original DataFrame head:")
display(df.head())

# Create 'DiscountedPrice' column (10% discount)
df['DiscountedPrice'] = df['Price'] * 0.9

# Create 'ProfitMargin' column (assuming cost is 70% of DiscountedPrice)
df['ProfitMargin'] = df['DiscountedPrice'] - (df['DiscountedPrice'] * 0.7)

print("\nDataFrame after adding 'DiscountedPrice' and 'ProfitMargin' columns:")
display(df.head())
```
